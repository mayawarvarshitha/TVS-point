### Model Building

```python
df.drop(['Education_det','Marital_Stat','Gen','Experience','Ph.No','Day_Process','Region','hour_Process','density_Ind'],axis=1,inplace=True)
```

```python
x=df.drop('Default',1)
y=df['Default']
df_d=pd.get_dummies(x,drop_first=True)
df_d.head()
```

```python
df_d['Default']=df['Default']
```

```python
x=df_d.drop(['Default'],axis=1)
y=df_d['Default']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3,random_state=100)
```

```python
LR=LogisticRegression()
model1=LR.fit(xtrain,ytrain)
```

```python
test=pd.read_csv('Test_Set_TVS.csv')
test.head(2)
```

```python
for i in test.select_dtypes(include='object'):
    test[i].fillna(test[i].mode()[0],inplace=True)
```

```python
for i in test.select_dtypes(exclude='object'):
    test[i].fillna(test[i].median(),inplace=True)
```

```python
test.drop(['ID','Education_det','Marital_Stat','Gen','Experience','Ph.No','Day_Process','Region','hour_Process','density_Ind'],axis=1,inplace=True)
```

```python
test=pd.get_dummies(test,drop_first=True)
test.head()
```

```python
ypred1=model1.predict(test)
```

```python
k=pd.read_csv('Sample_Submission_TVS.csv')
k.head(2)
```

```python
k['Default']=ypred1
```

```python
k.to_csv('sub1.csv')
```

```python
k.set_index('ID',inplace=True)
```

```python
SS=StandardScaler()
xtrain[df_d.iloc[:,0:10].columns]=SS.fit_transform(xtrain[df_d.iloc[:,0:10].columns])
xtest[df_d.iloc[:,0:10].columns]=SS.transform(xtest[df_d.iloc[:,0:10].columns])
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
PT=PowerTransformer()
xtrain[df_d.iloc[:,0:10].columns]=PT.fit_transform(xtrain[df_d.iloc[:,0:10].columns])
xtest[df_d.iloc[:,0:10].columns]=PT.transform(xtest[df_d.iloc[:,0:10].columns])
```

```python
KNN=KNeighborsClassifier()
model2=KNN.fit(xtrain,ytrain)
```

```python
ypred2=model2.predict(test)
k['Default']=ypred2
k.to_csv('sub2.csv')
```

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
model3=DT.fit(xtrain,ytrain)
```

```python
ypred3=model3.predict(test)
k['Default']=ypred3
k.to_csv('sub3.csv')
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
model4=RF.fit(xtrain,ytrain)
ypred4=model4.predict(test)
k['Default']=ypred4
k.to_csv('sub4.csv')
```

```python
AD=AdaBoostClassifier()
model5=AD.fit(xtrain,ytrain)
ypred5=model5.predict(test)
k['Default']=ypred5
k.to_csv('sub5.csv')
```

```python
GB=GradientBoostingClassifier()
model6=GB.fit(xtrain,ytrain)
ypred6=model6.predict(test)
k['Default']=ypred6
k.to_csv('sub6.csv')
```

```python
DT=DecisionTreeClassifier(criterion= 'entropy',max_depth=4,max_leaf_nodes= 8,
                          min_samples_leaf= 1,min_samples_split=2,random_state=10)
model7=DT.fit(xtrain,ytrain)
ypred7=model7.predict(test)
k['Default']=ypred7
k.to_csv('sub7.csv')
```

```python
XGRF=XGBRFClassifier()
model8=XGRF.fit(xtrain,ytrain)
ypred8=model8.predict(test)
k['Default']=ypred8
k.to_csv('sub8.csv')
```

```python
estimators = [('RF', RandomForestClassifier()),
             ('XG', XGBClassifier())]
STA = StackingClassifier(estimators=estimators, final_estimator=RandomForestClassifier())
model9=STA.fit(xtrain,ytrain)
ypred9=model9.predict(test)
k['Default']=ypred9
k.to_csv('sub9.csv')
```
